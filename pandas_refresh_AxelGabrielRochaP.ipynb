# Eliminar filas o columnas

La funcion `df.drop( fila/columna, axis=0/1, inplace=True/False )` sirve para eliminar filas o columnas. El primer par√°metro es la `fila` o `columna` que queremos eliminar. El segundo es el `axis`, puede ser `0` (para filas) o `1` (para columnas). El tercer par√°metro inplace puede ser `True` (si la modificaci√≥n se har√° directo sobre `df`) o `False` (si no). En caso de que no, `drop` regresar√° un nuevo dataframe modificado.


```python
# usa drop para eliminar la fila que tenga la posicion 1, directo sobre df
# y usa head para mostrar como qued√≥ df
datos_juegos = datos_juegos.drop(index=[1], axis=0)
datos_juegos.head()
```

```python
# Elimina la columna 3, pero el resultado generando el nuevo dataframe df2 y no en df
# y muestra df2 con head
df2= datos_juegos.drop(columns= [3], axis = [1])
df2.head()
```

```python
# muestra df con head
datos_juegos.head()
```

# Renombrar columnas

Nota que las columnas se llaman `0, 1, 2`  etc. La fila 0 tiene los nombres apropiados: `Name, Platform, Year_of_Release,` etc. Podemos asignar a `df.columns` el valor de toda la primera fila, y luego elimarla.

```python
# cambia el nombre de las columnas usando la fila 0, luego elimina esta fila,
# y finalmente muestra el resultado con la funci√≥n head.


# Cuidado si ejecutas esta celda varias veces, seguiras eliminado filas
dj3 = datos_juegos.copy()
dj3.columns = dj3.iloc[0]
dj3.drop(index=dj3.index[0], inplace=True)

dj3.head()

```

Tambi√©n se puede establecer que el nombre de las columnas esta en la primera fila desde el principio, al abrir el dataframe usando la opcion `header=0`, que es la opci√≥n por omisi√≥n.

```python
# vuelve a cargar el dataset en df usando los nombres de columna de la fila 0
# y muestra el resultado con head
df3= pd.read_csv('Gamess.csv')

df3.head()
```

# Manejar columnas

Podemos acceder a columnas especificas por nombre, por ejemplo `df.Name` . El resultado ser√° otro dataframe. Tambi√©n puedes seleccionar m√∫ltiples columnas a la vez usando una lista, por ejemplo `df[['Name', 'Series']]`.






```python
# crear un nuevo dataframe en la variable res, que solo contenga las columnas
# Name y Platform de df. Muestra el resultado con head
df4 = df3[['Name', 'Platform']]
df4.head()
```

Los nombres de las columnas tambi√©n se pueden usar para eliminarlas con `drop`.  Elimina la columna Global_Sales de df y muestra el resultado en otro dataframe.

```python
df3 = df3.drop(columns=['Global_Sales'])
df3.head()
```

# Agregar columnas nuevas

Para agregar columnas nuevas basta con asignarlas. Antes eliminaste la columna Global_Sales. Recalc√∫lala y agr√©gala de nuevo de nuevo a partir de la suma de NA_Sales, EU_Sales, JP_Sales y Other_Sales. Muestra el resultado con `df.head()`

```python
Global_Sales = df3['NA_Sales'] + df3['EU_Sales'] + df3['JP_Sales'] + df3['Other_Sales']
df3['Global_Sales'] = Global_Sales
df3.head()
```

# Manejo de NaN

Algunas columnas tienen valores faltantes (NaN). Por ejemplo, la columna **Critic_Score** tiene celdas con `NaN`. Averigua cu√°les datos de **Critic_Score** usando la funci√≥n `isna`. El resultado es una serie de booleanos, donde `True` indica cu√°les son `NaN`.  La funci√≥n `isnull` es similar a `isna`.

```python
# prompt: Algunas columnas tienen valores faltantes (NaN). Por ejemplo, la columna Critic_Score tiene celdas con NaN. Averigua cu√°les datos de Critic_Score usando la funci√≥n isna. El resultado es una serie de booleanos, donde True indica cu√°les son NaN. La funci√≥n isnull es similar a isna.

df3['Critic_Score'].isna()
```

Es usual elminar los NaN.  La funci√≥n `drop` puede hacerlo si recibe los √≠ndices que cumplen la condici√≥n de ser NaN. ¬°Hazlo sobre df, obteniendo los √≠ndices de las filas que tienen NaN en Critic_Score por medio de index!

```python
df_filtrado = df3.dropna(subset=['Critic_Score'])
df_filtrado
```

La funcion `fillna` sustituye los `NaN` por otra cosa que se indique. Esta es su sintaxis: `df['columna'] = df['columna'].fillna('valor sustituto')`   

La funcion mean sirve para obtener el promedio. Se puede usar sobre una columna. Por default, `mean` omitir√° los NaN que existan. Calcula el promedio de la columna NA_Sales.

```python
df3['NA_Sales'].mean()
```

# Funciones estad√≠sticas

Otras funciones como median, min, max tambi√©n est√°n disponibles. Si usas el argumento Numeric_only=True, se aplicar√≠an a todas las columnas num√©ricas.

```python
# Calcula el promedio, la mediana, el valor m√°ximo y el valor m√≠nimo para
# todas las columnas num√©ricas.
a = df3.mean(numeric_only=True)
b = df3.median(numeric_only=True)
c = df3.max(numeric_only=True)
d = df3.min(numeric_only=True)
print(a)
print(b)
print(c)
print(d)
```

# Ordenamiento

La funci√≥n `sort_values` permite ordenar un dataframe por el valor de una columna. Se usa inplace para determinar si se genera un nuevo dataframe o no.

```python
# ordenar df por el valor de la columna NA_Sales y mostrarlo con display
df3.sort_values(by='NA_Sales', inplace=True)
display(df3)


# sort_values tiene otras opciones que puedes consultar aqui
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html
```
